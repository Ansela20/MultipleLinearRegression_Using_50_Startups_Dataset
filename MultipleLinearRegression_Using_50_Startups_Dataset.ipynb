```python
# Multiple Linear Regression

#Importing the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# importing the Dataset
address = 'C:/Users/Ansela/Desktop/DataSet/50_Startups.csv'
Data = pd.read_csv(address)
Data.head()
```

```python
# Splitting the data based on independent(x) and dependent(y) variables
x = Data.iloc[:,:-1].values
X = pd.DataFrame(x)
X.columns = ['R&D Spend','Administration','Marketing Spend','State']
X.head()
```

```python
y= Data.iloc[:,4].values
Y = pd.DataFrame(y)
Y.columns = ['Profit']
Y.head()
```

```python
# Convert the column into categorical columns
states = pd.get_dummies(X['State'], drop_first= True)
states.head(10)
```

```python
# Drop the State Column
X = X.drop('State', axis =1)
X.head()
```

```python
#Concat the dummy variable
X = pd.concat([X, states], axis = 1)
X.head()
```

```python
# Dividing the complete dataset into training and test dataset
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.2, random_state = 0) #test_size = 0.2 implies 20% test and 80% train
X_train,X_test,Y_train,Y_test
```

```python
# Fitting Multiple Linear Regression to the Training Set
from sklearn.linear_model import LinearRegression
MultipleLinearRegression = LinearRegression() #classifier
MultipleLinearRegression.fit(X_train,Y_train)
```

```python
#Predicting the Test set results
Y_predict = MultipleLinearRegression.predict(X_test)
pd.DataFrame(Y_predict)
```

```python
from sklearn.metrics import r2_score
score = r2_score(Y_test, Y_predict)
score
```
